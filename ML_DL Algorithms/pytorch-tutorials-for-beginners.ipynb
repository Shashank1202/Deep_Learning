```python
x=torch.tensor([1, 2, 3])
y= torch.tensor([9,8,7])

z1=torch.empty(3)
print(torch.add(x, y,out=z1))

print(x+y)

print(x-y)

print(torch.true_divide(x, y))

print(x*y)
```

```python
t=torch.zeros(3)
t.add_(x)
print(t)
```

```python
z=x.pow(2)
print(z)

print(x**2)
```

```python
#Matrix Multiplication

x1=torch.rand((2, 5))
x2=torch.rand((5, 3))
x3= torch.mm(x1, x2)
print(x1)
print(x2)
print(x3)
```

```python
matrix_exp= torch.rand(5,5)
print(matrix_exp.matrix_power(3))
```

```python
z= torch.dot(x, y)
print(z)
```

```python
batch=32
n=10
m=20
p=30

tensor1= torch.rand((batch, n, m))
tensor2= torch.rand((batch,m,p))
out_bmm= torch.bmm(tensor1, tensor2)
print(out_bmm)
```

```python
# More Advanced indexing

x= torch.arange(10)
print(x[(x<2) |(x>8)])
```

```python
#Tensor Reshaping

x=torch.arange(9)

x_3X3= x.view(3, 3)
x_3X3= x.reshape(3, 3)
print(x_3X3)
```

```python
y= x_3X3.t()
print(y)
```

```python
print(y.contiguous().view(9))
```

# Importing the Required Libraries

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

# Let's Load the data into Training and Testing data.

```python
(x_train, y_train), (x_test, y_test)=mnist.load_data()
```

# Let's Reshape the data with height, width and Channels

```python
x_train= x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test= x_test.reshape(x_test.shape[0], 28, 28, 1)
```

# Here We are normalizing the data and converting into float standard format

```python
x_train= x_train.astype('float32')/255.0
y_train= y_train.astype('float32')/255.0
```

# One-hot Encoding the labels

```python
y_train= to_categorical(y_train, 10)
y_test= to_categorical(y_test, 10)
```

# Data Augmentation

```python
datagen= ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1
)

datagen.fit(x_train)
```

# Building the CNN Models with layers like(Conv2D, MaxPooling, Dense layer) Finally Flattening the data to 1D

```python
model=Sequential()

model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))

model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))

model.add(Dropout(0.5))

model.add(Dense(10, activation='softmax'))
```

# Compiling the Model

```python
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

# Let's get to the final part:: That's Training the Model

```python
model.fit(datagen.flow(x_train, y_train, batch_size=100),
         validation_data=(x_test, y_test),
         epochs=10)
```

# Evaluation of the Model.

```python
score= model.evaluate(x_test, y_test, verbose=0)
print("Test Loss:", score[0])
print("Test accuracy:",score[1])
```

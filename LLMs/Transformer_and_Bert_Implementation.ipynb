```python
from transformers import pipeline
classifier= pipeline('sentiment-analysis')
```

```python
classifier("I'm very happy to learn new things everyday")
```

```python
classifier("Worst hotels, all old food")
```

```python
results= classifier(["We are very happy to serve you",
                     "We hope you don't hate our service"])

for result in results:
  print(f"Label: {result['label']}, with score: {round(result['score'], 2)}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name= "nlptown/bert-base-multilingual-uncased-sentiment"

model= TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt= True)
tokenizer= AutoTokenizer.from_pretrained(model_name)
classifier= pipeline('sentiment-analysis', model= model, tokenizer= tokenizer)
```

```python
classifier("Iam a good boy")
```

```python
input= tokenizer("I'm very happy to learn new things everyday", return_tensors= "tf")
```

```python
print(input)
```

```python
tf_batch= tokenizer(
    ["I'm very happy to learn new things everyday",
     "I'm very sad to learn new things everyday"],
     padding= True,
     truncation= True,
     max_length= 512,
     return_tensors= "tf"
)
```

```python
for key, value in tf_batch.items():
  print(f"{key}: {value.numpy().tolist()}")
```
